## Cleansing
### receipttime, settime, ridetime
- 동일한 datetimeindex로 자료형 변환
- 각 차이 시간 계산한 새로운 칼럼 생성

```python
import datetime as dt
```

```python
df_reiptime_pm = df[df['receipttime'].str.contains('오후', na = False)]

# 오후 1시 이후로 호출시간이 적제 되었는지 체크 ✒︎ 오후 1시까지만 있음 확인
df_reiptime_pm[df_reiptime_pm['receipttime'].str.contains('오후 1:', na = False)] 
df_reiptime_pm[df_reiptime_pm['receipttime'].str.contains('오후 1:', na = False)] ## 오전12시 ✒︎ 밤12시 | 오후12시 ✒︎ 낮12시
```

```python
# '오전','오후' 명칭 없애고 24시간제로 변환하기
df['am'] = pd.to_datetime(df[df['ridetime'].str.contains('오전',na = False)]['ridetime'].str.replace('오전',''))
df['pm'] = pd.to_datetime(df[df['ridetime'].str.contains('오후',na = False)]['ridetime'].str.replace('오후','')) + dt.timedelta(hours = 12)
df.head()
```

null값 확인하기

```python
print(df['am'].isnull().sum())
print(df['pm'].isnull().sum())
```

```python
df[df['pm'].notnull()] 
```

```python
print(len(df[df['am'].isnull() == df['pm'].isnull()]))
df[df['am'].isnull() == df['pm'].isnull()] ## ridetime : null값 인값 존재 -> 차량 정보 누락 10행 제거
```
