```python
import pandas as pd
import numpy as np
import datetime  as dt
import seaborn as sns
import glob
```

```python
df= pd.read_csv('master_final_220708_택시요금추가.csv')
df.head()
```

```python
df.sort_values(by = '배차시간')
```

```python
df["예정시간"]=pd.to_datetime(df["예정시간"])
df['배차시간']= pd.to_datetime(df['배차시간'])
df["승차시간"]=pd.to_datetime(df["승차시간"])


# df['배차시간_y'] = df['배차시간'].dt.year
# df['배차시간_h'] = df['배차시간'].dt.hour

df['배차시간_yh'] = df['배차시간'].dt.strftime('%Y%m')
```

```python
df.iloc[:,-7:].head()
```

```python
#기준년월, 구 , Taxifare만 남기기 
df_new = df[['배차시간_yh','startpos1','taxiFare_I']]
df_new
```

```python
df_new = df_new.groupby(['startpos1','배차시간_yh']).sum()["taxiFare_I"].reset_index()
df_new.rename(columns = {'startpos1':'구','배차시간_yh':'기준년월','taxiFare_I':'택시요금'},inplace=True)
# df_new = df_new[df_new['구'] in ('강서구','강북구','노원구','서대문구','강동구','은평구')]
# csv로 저장합니다.
df_new.to_csv('master_final_220711_구별택시요금.csv', index= False, encoding = 'utf-8')
```
