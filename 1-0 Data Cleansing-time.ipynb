## Cleansing
### receipttime, settime, ridetime
- 동일한 datetimeindex로 자료형 변환
- 각 차이 시간 계산한 새로운 칼럼 생성

```python
import datetime as dt
```

```python
df_reiptime_pm = df[df['receipttime'].str.contains('오후', na = False)]

# 오후 1시 이후로 호출시간이 적제 되었는지 체크 ✒︎ 오후 1시까지만 있음 확인
df_reiptime_pm[df_reiptime_pm['receipttime'].str.contains('오후 1:', na = False)] 
df_reiptime_pm[df_reiptime_pm['receipttime'].str.contains('오후 1:', na = False)] ## 오전12시 ✒︎ 밤12시 | 오후12시 ✒︎ 낮12시
```

### Ridetime 탑승시간

```python
# '오전','오후' 명칭 없애고 24시간제로 변환하기
df['am'] = pd.to_datetime(df[df['ridetime'].str.contains('오전',na = False)]['ridetime'].str.replace('오전',''))
df['pm'] = pd.to_datetime(df[df['ridetime'].str.contains('오후',na = False)]['ridetime'].str.replace('오후','')) + dt.timedelta(hours = 12)
df.head()
```

```python
df.info()
```

null값 확인하기

```python
print(df['am'].isnull().sum())
print(df['pm'].isnull().sum())
```

```python
df[df['pm'].notnull()] 
```

am, pm 둘다 null값 존재 확인

```python
print(len(df[df['am'].isnull() == df['pm'].isnull()]))
df[df['am'].isnull() == df['pm'].isnull()] ## ridetime null값 인값 존재 -> 차량 정보 누락 10행 제거
```

```python
df = df.dropna(subset=['am', 'pm'], how='all') 
df #407613 -> 407603 
```

am, pm 합치기

```python
# df['am'][0] + df['pm'][0] ## NaT는 연산이 불가능 하므로 0으로 대체 
```

```python
dt.timedelta(hours=0)
```

```python
df['am'] = df['am'].fillna(dt.timedelta(hours=0))
df['pm'] = df['pm'].fillna(dt.timedelta(hours=0))
```

```python
df['am'][0] + df['pm'][0] #timestamp로 합쳐지는 것 확인 
```

```python
%%time

승차_val = []
for i in range(len(df)):
    try: 
        val = df['am'][i] + df['pm'][i]
        승차_val.append(val)
    except:
        승차_val.append(dt.timedelta(hours=0))
```

```python
df['승차시간'] = 승차_val
```

```python
df[df['승차시간'] == dt.timedelta(hours=0)] ##ridetime 버젓이 있는데 승차시간 빵인건 무슨 경우??
```

```python
con = df['승차시간'] == dt.timedelta(hours=0)
df[con]['승차시간'] = pd.to_datetime(df[con]['am'])

sindx = df[con]['승차시간'].index

df.loc[sindx, '승차시간'] = pd.to_datetime(df[con]['am'])

df[df['승차시간'] == dt.timedelta(hours=0)]

df['승차시간'] = pd.to_datetime(df['승차시간'])

df = df.drop(['am','pm'], axis = 1)

df.info()
```

### settime 배차시간

```python
df['settime'].str.contains('오후')

df[df['settime'].isna()].head()

df_set_na = df[df['settime'].isna()]

snax = df[df['settime'].isna()].index

df = df.drop(snax)

df['오전'] = pd.to_datetime(df[df['settime'].str.contains('오전')]['settime'].str.replace('오전', ''))
df['오후'] = pd.to_datetime(df[df['settime'].str.contains('오후')]['settime'].str.replace('오후', '')) + dt.timedelta(hours = 12)

df['오후'] = df['오후'].fillna(dt.timedelta(hours=0))
df['오전'] = df['오전'].fillna(dt.timedelta(hours=0))
```

```python
%%time

승차_val = []
for i in range(len(df)):
    try: 
        val = df['오후'][i] + df['오전'][i]
        승차_val.append(val)
    except:
        승차_val.append(dt.timedelta(hours=0))
```

```python
df['배차시간'] = 승차_val
df[df['배차시간'] == dt.timedelta(hours=0)].head(2)
```

```python
con = df['배차시간'] == dt.timedelta(hours=0)

binx = df[con]['배차시간'].index

df.loc[binx, '배차시간'] = pd.to_datetime(df[con]['오전']) 
```

```python
df['배차시간'] =pd.to_datetime(df['배차시간'])
df.info()
```

```python
df = df.drop(['오전', '오후'], axis=1)
```

### receipttime 예정시간

```python
df[df['receipttime'].isna()]

df['오전'] = pd.to_datetime(df[df['receipttime'].str.contains('오전')]['receipttime'].str.replace('오전', ''))
df['오후'] = pd.to_datetime(df[df['receipttime'].str.contains('오후')]['receipttime'].str.replace('오후', ''))

df['오후'] = df['오후'].fillna(dt.timedelta(hours=0))
df['오전'] = df['오전'].fillna(dt.timedelta(hours=0))
```

```python
%%time

요청_val = []
for i in range(len(df)):
    try: 
        val = df['오후'][i] + df['오전'][i]
        요청_val.append(val)
    except:
        요청_val.append(dt.timedelta(hours=0))
```

```python
df['예정시간'] = 요청_val

df[df['예정시간'] == dt.timedelta(hours=0)]['receipttime']

con = df[df['예정시간'] == dt.timedelta(hours=0)]['receipttime'].str.contains('오전')
con1 = df[df['예정시간'] == dt.timedelta(hours=0)]['receipttime'].str.contains('오후')

con1[con1 == True]
einx = con[con == True].index
df.loc[einx, '예정시간'] = df.loc[einx, '오전']
df.loc[einx, '예정시간']
```

```python
df_re_date = df[df['예정시간'] == dt.timedelta(hours=0)]
df_re_date.to_csv('df_re_date.csv', index = False, encoding = 'utf-8')
dinx = df[df['예정시간'] == dt.timedelta(hours=0)].index
df = df.drop(dinx)
df = df.drop(['오전', '오후'], axis=1)

df_cleansing2 = df[['no', 'cartype', '예정시간', '배차시간', '승차시간', 'startpos1', 'startpos2', 'endpos1', 'endpos2']]
df_cleansing2.info()
```

```python
df_cleansing2['예정시간'] = pd.to_datetime(df_cleansing2['예정시간'])
df_cleansing2.info()
```

인덱스 정렬

```python
df_cleansing2 = df_cleansing2.sort_index(ascending= True)
df_cleansing2 = df_cleansing2.reset_index(drop=True)
df_cleansing2.to_csv('master_cleansing_space_time.csv', index = False, encoding = 'utf-8')
```
